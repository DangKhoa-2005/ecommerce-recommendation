```python
# Khai báo đường dẫn file từ Volume
file_path = "/Volumes/workspace/default/ecommerce_raw/Brazilian E-Commerce Public Dataset by Olist.csv"

# Đọc dữ liệu từ file CSV (Data Ingestion - Bronze Layer)
df_raw = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)

# Hiển thị 10 dòng đầu tiên ra màn hình
display(df_raw.limit(10))
```

```python
# Xử lý làm sạch dữ liệu cơ bản (ETL - Silver Layer)
# Loại bỏ các dòng trùng lặp và các dòng trống hoàn toàn
df_clean = df_raw.dropDuplicates().dropna(how="all")

# Định nghĩa đường dẫn lưu Delta Lake cũng nằm trong Volume bạn vừa tạo
delta_table_path = "/Volumes/workspace/default/ecommerce_raw/ecommerce_silver"

# Ghi dữ liệu xuống định dạng Delta table tiêu chuẩn
df_clean.write \
    .format("delta") \
    .mode("overwrite") \
    .save(delta_table_path)

# Truy vấn lịch sử table để lấy minh chứng Delta Lake
display(spark.sql(f"DESCRIBE HISTORY delta.`{delta_table_path}`"))
```
